```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data for plotting
t = np.arange(0.1, 2.0, 0.01)
s = 1/(np.exp(40*t-30)+1)
```

```python
fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Distribuição de Fermi-Dirac')

# fig.savefig("test.png")
plt.show()
```
