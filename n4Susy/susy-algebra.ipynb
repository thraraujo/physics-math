### N=4 SYM & psu(2,2|4) Algebra

Here I want to review some aspects of N=4 SYM and its algebra. I want to follow Beisert et. al. [big review on integrability](https://arxiv.org/abs/1012.3982).

```python
from sympy import MatrixSymbol, Matrix # Sympy allows us to perform some algebraic manipulations for matrices. 
```

Let me now create some matrices. 

```python
A, B, C, D = MatrixSymbol('A',4,4), MatrixSymbol('B',4,4), MatrixSymbol('C',4,4), MatrixSymbol('D',4,4)
```

```python
E, F, G, H = MatrixSymbol('E',4,4), MatrixSymbol('F',4,4), MatrixSymbol('G',4,4), MatrixSymbol('H',4,4)
```

```python
X = Matrix([[A,B],[C,D]])
```

```python
Y = Matrix([[E,F],[G,H]])
```

```python
def comm(a,b):
    return a*b - b*a
```

```python
comm(X,Y)
```
