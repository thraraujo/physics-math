# Streamlines 

Aqui queremos plotar algumas streamlines para o movimento de alguns fluidos. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Antes de mais nada, quero definir algumas funções que podemos modificar posteriormente


```python
w = 10
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = -0.5 * X
V = 0.5 * Y

fig, ax = plt.subplots(figsize=(4, 4))

ax.streamplot(X, Y, U, V, color='cornflowerblue', density = 1.3)
```

```python
w = 10
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = -0.5 * Y
V = 0.5 * X

fig, ax = plt.subplots(figsize=(4, 4))

ax.streamplot(X, Y, U, V, color='cornflowerblue', density = 1)
```
