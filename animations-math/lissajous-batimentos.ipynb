# Curvas de Lissajous e Batimentos

# Batimentos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def x(x, freq, amp = 1, fase = 0):
    return amp * np.cos(freq * x + fase)
```

```python
t = np.arange(0,50, 0.01)
```

```python
plt.plot(t, x(t,np.sqrt(5)) + x(t,4))
```

# Lissajous

```python
plt.plot(x(t,np.pi), x(t,np.e))
```
